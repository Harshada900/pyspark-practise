```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as f

spark = SparkSession.builder \
    .appName("Topic10") \
    .config("spark.sql.shuffle.partitions", "5") \
    .getOrCreate()

data = [
    (1, "Raj Sharma",    "raj@gmail.com",     "+91-9876543210", "2024/01/15"),
    (2, "Priya Mehta",   "PRIYA@YAHOO.COM",   "+91-8765432109", "2024/02/20"),
    (3, "Amit Kumar",    "amit@hotmail.com",  "+91-7654321098", "2024/03/25"),
    (4, "Sara Khan",     "SARA@GMAIL.COM",    "+91-6543210987", "2024/04/30"),
    (5, "Karan   Singh", "karan@outlook.com", "+91-5432109876", "2024/05/15")
]

columns = ["id", "full_name", "email", "phone", "join_date"]
df = spark.createDataFrame(data, columns)
df.show(truncate=False)
```

## **1. Case Functions — upper, lower, initcap**

```python
#uppercase

df.withColumn("full_name_upper", f.upper(f.col("full_name"))).show(truncate=False) #truncate=False means it'll show full string
```

```python
df.withColumn("email", f.lower(f.col("email"))).show(truncate=False)
```

```python
df.withColumn("full_name_initcap", f.initcap(f.col("full_name"))).show(truncate=False)
```

## **2. concat — Joining Strings Together and concat_ws - concatinates strings but skips nulls**

```python
df.withColumn("info", f.concat(f.col("full_name"), f.lit(" | "), f.col("email"))).show(truncate=False)
```

```python
#concat_ws - concat with separator

df.withColumn("info", f.concat_ws(" | ", f.col("full_name"), f.col("email"))).show(truncate=False)
```

```python
from pyspark.sql.types import StructType, StructField, StringType

# Example showing the difference
data2 = [("Raj", None, "India")]

# Define explicit schema to handle None values
schema = StructType([
    StructField("name", StringType(), True),
    StructField("email", StringType(), True), # Specify StringType for email
    StructField("country", StringType(), True)
])

df2 = spark.createDataFrame(data2, schema)

df2.withColumn("test1", f.concat(f.col("name"), f.lit("-"), f.col("email"))).show()
# Output: null  ← because email is null

df2.withColumn("test2", f.concat_ws("-", f.col("name"), f.col("email"), f.col("country"))).show()
# Output: Raj-India  ← skipped null email
```

## **3. split — Splitting a String into Array**

```python
df.withColumn("name_array", f.split(f.col("full_name"), " ")).show(truncate=False)
```

```python
#extracting first and last name

df.withColumn("first_name", f.split(f.col("full_name"), " ").getItem(0)).withColumn("last_name", f.split(f.col("full_name"), " ").getItem(1)).show(truncate=False)
```

## **4. regexp_replace — Replace Using Pattern**

.    → any single character

\d   → any digit (0-9)

\s   → any whitespace

(+)    → one or more of previous

(*)    → zero or more of previous

^    → start of string

$    → end of string

```python
#Clean Phone Numbers

df.withColumn("cleaned_phone", f.regexp_replace(f.col("phone"), "[^0-9]", "")).show(truncate=False)
```

```python
#remove +91-

df.withColumn("cleaned_phone", f.regexp_replace(f.col("phone"),  r"\+91-", "")).show(truncate=False)
```

```python
#Fix Date Format - replace / with -

df.withColumn("join_date", f.regexp_replace(f.col("join_date"), "/", "-")).show(truncate=False)
```

```python

#Remove Extra Spaces

df.withColumn("cleaned_name", f.regexp_replace(f.col("full_name"), r"\s+", " ")).show(truncate=False)
```

## **5. Other Commonly Used String Functions**

### 1 trim

```python
#remove spaces from both side left and right

df.withColumn("full_name", f.trim(f.col("full_name"))).show(truncate=False)
```

```python
#remove space from left side only

df.withColumn("full_name", f.ltrim(f.col("full_name"))).show(truncate=False)
```

```python
#remove space from right side only
df.withColumn("full_name", f.rtrim (f.col("full_name"))).show(truncate=False)
```

# length — Get string length

```python
df.withColumn("name_length", f.length(f.col("full_name"))).show(truncate=False)
```

# substring — Extract part of string

```python
df.withColumn("year", f.substring(f.col("join_date"), 1, 4)).show(truncate=False)   # start at position 1, take 4 chars
```

```python
df.withColumn("date", f.substring(f.col("join_date"), 9, 10)).show(truncate=False) #PySpark substring starts at position 1 not 0! Unlike Python which starts at 0
```

## contains, startsWith, endsWith

```python
#filter emails containing gmail

df.filter(f.col("email").contains("gmail")).show(truncate=False)
```

```python
#filter names that starts with R

df.filter(f.col("full_name").startswith("R")).show(truncate=False)
```

```python
#filter emails that ends with .com

df.filter(f.col("email").endswith(".com")).show(truncate=False)
```

## like — SQL style pattern matching

```python
#emails that contains gmail

df.filter(f.col("email").like("%gmail%")).show(truncate=False)
```

```python
#name starts with S

df.filter(f.col("full_name").like("S%")).show(truncate=False)
```

```python
#name ends with a

df.filter(f.col("full_name").like("%a")).show()
```

```python
df.withColumn("split_names", f.split(f.col("full_name"), " ")).show()   #gives empty string for whitespaces!
# "Karan   Singh" splits into ["Karan", "", "", "Singh"]
# getItem(1) gives "" (empty string) not "Singh"!)
```
