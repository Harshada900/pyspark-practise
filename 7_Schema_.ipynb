## **StructField -> Defines One Column**

```python
from pyspark.sql.types import StructField, StringType, IntegerType, StructType, DoubleType, LongType, BooleanType, DateType

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Struct").getOrCreate()
```

```python
#syntax

StructField("name", StringType(), True)
```

## **StructType - collection of StructField**

```python
schema = StructType([
    StructField("emp_id", StringType(), False),
    StructField("name", StringType(), True),
    StructField("country", StringType(), True),
    StructField("department", StringType(), True),
    StructField("salary", DoubleType(), True)
])
```

## **Using Schema While Reading**

```python
#df = spark.read.csv("file.csv", header=True, schema=schema) # note: no file.csv..
```

## **Using Schema While Creating DataFrame**

```python
data = [(1, "Raj",   "India", "Engineering", 50000.0),
        (2, "Priya", "USA",   "Marketing",   70000.0),
        (3, "Amit",  "India", "Engineering", 60000.0)]

df = spark.createDataFrame(data, schema)
df.show()
```

```python
df.printSchema()
```

## **DDL String Format — Shortcut Way**

```python
schema_ddl = "emp_id INT, name STRING, country STRING, department STRING, salary DOUBLE"
df_ddl = spark.createDataFrame(data, schema_ddl)
df_ddl.show()
```

```python
df_ddl.printSchema()
```

## **Getting Schema of Existing DataFrame**

```python
# Print schema in readable format
df.printSchema()
```

```python
# Get schema as StructType object (useful for reusing)
print(df.schema)
```

```python
# Get just column names and types
print(df.dtypes)
```

## **Nested Schema — StructType Inside StructType**

```python
# JSON data like: {"name": "Raj", "address": {"city": "Mumbai", "pin": 400001}}

nested_schema = StructType([
  StructField("name", StringType(), True),
  StructField("address", StructType([
    StructField("city", StringType(), True),
    StructField("pin", IntegerType(), True)
  ]), True)
])
```

```python
nested_data = [("Raj",   {"city": "Mumbai",    "pin": 400001}),
        ("Priya", {"city": "Delhi",     "pin": 110001})]
```

```python
nested_df= spark.createDataFrame(nested_data, nested_schema)
nested_df.show()
```

```python
nested_df.printSchema()
```

Accessing nested columns:

```python
nested_df.select("name", "address.city").show() # use dot notation
```

## **ArrayType — Column with Multiple Values**

```python
from pyspark.sql.types import ArrayType
```

```python
arr_schema = StructType([
    StructField("name", StringType(), True),
    StructField("languages", ArrayType(StringType()), True)
])
```

```python
arr_data = [("Raj",   ["Python", "Spark", "SQL"]),
        ("Priya", ["Excel",  "Power BI"])
        ]
```

```python
arr_df = spark.createDataFrame(arr_data, schema=arr_schema) #pysparks array can not contain mixed types !
arr_df.show()
```

```python
arr_df.printSchema()
```
