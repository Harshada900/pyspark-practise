```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('session1').config("spark.sql.shuffle.partitions", "5").getOrCreate()
```

```python
# Create sample data
data = [("Raj", "India", 50000),
        ("Priya", "USA", 70000),
        ("Amit", "India", 60000),
        ("Sara", "UK", 80000)]

columns = ["name", "country", "salary"]
```

```python
#creating a dataframe

df = spark.createDataFrame(data=data, schema=columns)
```

```python
#task 1: write as parquet with overwrite mode..

df.write.mode("overwrite").parquet("/content/output/data")
```

```python
#task 2 write partition by country

df.write.mode("overwrite").partitionBy("country").parquet("/content/output/partitioned")
```

```python
#task 3 check what files got created

import os

for root, dirs, files in os.walk("/content/output"):
  for file in files:
    print(os.path.join(root, file))
```
