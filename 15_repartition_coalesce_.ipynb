```python
from pyspark.sql import SparkSession

spark  = SparkSession.builder.appName("PartitionPractice").config("sparl.sql.shuffle.partirions", "5").getOrCreate()
```

```python
# Create DataFrame
data = [(i, f"name_{i}", i % 3) for i in range(1, 101)]
df = spark.createDataFrame(data, ["id", "name", "group"]) #df is created un memory python list
```

```python
df.show(9)
```

## **Check current partition count**

```python

print(df.rdd.getNumPartitions())# sspark decides partitions based on no of cores available on ur machine - google colab has 2
```

## **Repartition to 8**

```python
df8 = df.repartition(8)

print(df8.rdd.getNumPartitions())

df8.count()
```

## **Reduced it to 2**

```python
df2= df8.coalesce(2) #no shuffle only decreases , uneven data distribution possible

print(df2.rdd.getNumPartitions())

df2.count()
```
