```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

spark = SparkSession.builder.appName('session4').config("spark.sql.shuffle.partitions", "5").getOrCreate()
```

```python
data = [("Raj",   "India", "Engineering", 50000),
        ("Priya", "USA",   "Marketing",   70000),
        ("Amit",  "India", "Engineering", 60000),
        ("Sara",  "UK",    "HR",          80000),
        ("Karan", "India", "Marketing",   45000),
        ("John",  "USA",   "Engineering", 90000),
        ("Neha",  "India", "HR",          55000),
        ("Priya", "USA",   "Marketing",   70000)]  # duplicate row added

columns = ["name", "country", "department", "salary"]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
#orderBy - by default sorts in ascending (low to high or A to Z in case of string type)

df.orderBy("salary").show()
```

```python
from pyspark.sql.functions import col
```

```python
#descending

df.orderBy(col("salary").desc()).show()
```

```python
#shorthand way

df.orderBy(F.desc("salary")).show()
```

### **Sort By Multiple Columns**

```python
#sort country A - Z then sort salary high to low for each country

df.orderBy(col("country").asc(),
           col("salary").desc()
           ).show()
```

```python
#limit - returns top N rows as dataframe

df.limit(3).show()
```

```python
#find top 3 highest paid employees

df.orderBy(col("salary").desc()).limit(3).show()
```

```python
df.take(3)    #returns list of row objects - action not a transformation like limit
```

```python
df.head(3)    #returns list of row objects - action not a transformation like limit
```

```python
#distinct - removes duplicates where every single column value is identical/same

df.distinct().show()
```

```python
#dropDuplicates() - it lets you specify which columns to consider when checking duplicates
df.dropDuplicates(["name", "country"]).show()
```

```python
data = [("Raj", "India", 50000),
        ("Raj", "India", 60000),   # same name & country, different salary
        ("Priya", "USA", 70000),
        ("Priya", "USA", 70000)]   # completely duplicate

df2 = spark.createDataFrame(data, ["name", "country", "salary"])

# distinct() — keeps both Raj rows (salary is different so not fully duplicate)
df2.distinct().show()
# Output: 3 rows

# dropDuplicates on name & country — keeps only first Raj row
df2.dropDuplicates(["name", "country"]).show()
# Output: 2 rows
```
