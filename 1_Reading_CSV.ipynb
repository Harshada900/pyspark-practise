```python
# sample data
data = [("Raj", 25, 50000), ("Priya", 30, 70000), ("Amit", 28, 60000)]
columns = ["name", "age", "salary"]
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MyPractice") \
    .getOrCreate()
```

```python
print(spark.version)
```

```python
df = spark.createDataFrame(data, columns)
df.write.csv("sample_data.csv", header=True)
df.show()
```

```python
#readind data
df =spark.read.csv("sample_data.csv", header=True) # no data bcz of lazy evaluation
```
